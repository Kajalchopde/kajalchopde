```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

d1=sns.load_dataset('penguins')
d1.head(40)
glue=sns.load_dataset("glue").pivot(index="Model",columns="Task",values='Score')
iris=sns.load_dataset('iris')
sns.barplot(x='bill_length_mm',y='bill_depth_mm',data=d1,color='yellow',palette='rocket')
sns.displot(d1['bill_length_mm'])
sns.countplot(x='bill_depth_mm',hue='species',data=d1,palette='Set1')
sns.distplot(d1.bill_length_mm,bins=10,color='g')
sns.heatmap(glue,cbar=True,linewidths=0.5,annot=True,fmt=".1f",cmap="crest")
sns.scatterplot(x="sepal_length",y='petal_length',data=iris,hue='species')
sns.pairplot(iris,hue='species',palette='Set1')
plt.show
```
