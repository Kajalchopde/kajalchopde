```python
#it converts the column into an iterator
# the fuction that we will pass in the apply() is called each observation

#df.Gender.apply(lambda x: 1 if x == 'male' else 0) ('male')
df['Gender'] = df.Gender.apply(lambda x: 1 if x == 'male' else 0) 
```

```python
#map()
df['Pclass'] = df.Pclass.map(lambda x: 'First' if x == 1 else 0)
```

```python
#df('Gender')=np.where((df['Gender']))
df['Gender']=df.Gender.replace('male',0)
```
