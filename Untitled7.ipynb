```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
d1=sns.load_dataset('penguins')
d1=d1.head(10)
d2=pd.read_csv(r'./prac data/data.csv')
d2=d2.head(10)
```

```python
a1=sns.barplot(x='bill_length_mm',y='bill_depth_mm',
               data=d1, color='yellow',palette='rocket')
plt.show
```

```python
a2=sns.displot(d1['bill_length_mm'])
plt.show
```

```python
a3=sns.countplot(x='bill_depth_mm',hue='species',data=d1, palette='Set1')
plt.show()
```

```python
a4=sns.distplot(d1.bill_length_mm,bins=10,color='g')
plt.plot
```

```python
glue=sns.load_dataset("glue").pivot(index="Model",columns="Task",values="Score")
a5=sns.heatmap(glue,cbar=True,linewidths=0.5,annot=True,fmt=".1f",cmap="crest")
a5.set(xlabel="",ylabel="")
a5.xaxis.tick_top()
```

```python
iris=sns.load_dataset('iris')
iris.head()
a6=sns.scatterplot(x='sepal_length',y='petal_length',data=iris,hue='species')
plt.show
```

```python
a6=sns.pairplot(iris,hue='species',palette='Set1')
plt.show
```
