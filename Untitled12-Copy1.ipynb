```python
temp_cel=[42,34,28,34]

```

```python
t=34
(9/5)*t+32
```

```python
def ctemp(lst):
    e_lst=[]
    for i in lst:
        e_lst.append(((9/5)*i+32))
    return e_lst
```

```python
ctemp(temp_cel)
```

```python
def ct(lst):
    for i in lst:
        yield((9/5)*i+32)
```

```python
list(ct(temp_cel))
```

```python
def ctt(lst):
    return ((9/5)*lst+32)
```

```python
list(map(ctt,temp_cel))
```

```python
def k1(lst1):
    return(lst1**2)
```

```python
lst1=(1,2,3,4,5)
```

```python
def k2(lst1):
    for i in lst1:
        yield(i**2)
```

```python
list(k2(lst1))
```

```python
def k3(lstt):
    kk=[]
    for i in lstt:
        kk.append(i**2)
    return kk
        
```

```python
k3(lst1)
```

```python
return - return result and stop function process
yield - return result when loop complete
map - collect result till loop complete and at last it show
```

```python
def eo(l1):
    p1=[]
    for i in l1:
        if i%2==0:
            p1.append("Even,")
        else:
            p1.append("Odd,")
    return p1
```

```python
eo(col_1)
```

```python
def eo1(l1):
    for i in l1:
        if i%2==0:
            yield(("Even,"))
        else:
            yield(("Odd,"))
    
```

```python
list(eo1(col_1))
```

```python
def eo3(l1):
    if l1%2==0:
        return("Even,")
    else:
        return("Odd,")
    
```

```python
list(map(eo3,col_1))
```
